```python
import pandas as pd
import matplotlib.pyplot as plt

def find_pareto_front(df):
    # Sort by f1(x), which we'll assume is Total Cost
    df_sorted = df.sort_values('f1(x)')
    pareto_front = []
    max_quality = float('-inf')
    
    # Iterate through the sorted DataFrame to find the Pareto front
    for index, row in df_sorted.iterrows():
        if row['f2(x)'] > max_quality:
            pareto_front.append(row)
            max_quality = row['f2(x)']
    
    return pd.DataFrame(pareto_front)

def plot_pareto_front(csv_file):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Ensure the columns are correctly named
    if 'f1(x)' not in df.columns or 'f2(x)' not in df.columns:
        raise ValueError("CSV file must contain 'f1(x)' and 'f2(x)' columns")
    
    # Find the Pareto front
    pareto_df = find_pareto_front(df)
    
    # Plot the data points
    plt.scatter(df['f1(x)'], df['f2(x)'], label='Data Points')
    plt.plot(pareto_df['f1(x)'], pareto_df['f2(x)'], color='red', marker='o', label='Pareto Front')
    
    # Label the axes
    plt.xlabel('Total Cost')
    plt.ylabel('Quality Index')
    plt.title('Pareto Front Plot')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_pareto_front('/Users/vaniamiguel/PycharmProjects/MegaCon_final/MegaCon/f_output.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from paretoset import paretoset

def plot_pareto_front(csv_file):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Ensure the columns are correctly named
    if 'f1(x)' not in df.columns or 'f2(x)' not in df.columns:
        raise ValueError("CSV file must contain 'f1(x)' and 'f2(x)' columns")
    
    # Find the Pareto front using paretoset
    is_pareto = paretoset(df[['f1(x)', 'f2(x)']], sense=['min', 'max'])
    pareto_df = df[is_pareto]
    
    # Plot the data points
    plt.scatter(df['f1(x)'], df['f2(x)'], label='Data Points')
    plt.plot(pareto_df['f1(x)'], pareto_df['f2(x)'], color='red', marker='o', label='Pareto Front')
    
    # Label the axes
    plt.xlabel('Total Cost')
    plt.ylabel('Quality Index')
    plt.title('Pareto Front Plot')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_pareto_front('/Users/vaniamiguel/PycharmProjects/MegaCon_final/MegaCon/f_output.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('f_output.csv')

# Rename columns for clarity
df.columns = ['Total Cost', 'Quality Index']

# Function to identify Pareto-optimal points
def is_pareto_efficient(costs):
    is_efficient = np.ones(costs.shape[0], dtype = bool)
    for i, c in enumerate(costs):
        if is_efficient[i]:
            is_efficient[is_efficient] = np.any(costs[is_efficient] < c, axis=1)
            is_efficient[i] = True
    return is_efficient

# Identify Pareto-optimal points
costs = df[['Total Cost', 'Quality Index']].values
pareto_optimal = is_pareto_efficient(costs)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Total Cost'][~pareto_optimal], df['Quality Index'][~pareto_optimal], 
            alpha=0.6, label='Dominated Solutions')
plt.scatter(df['Total Cost'][pareto_optimal], df['Quality Index'][pareto_optimal], 
            color='red', alpha=1, label='Pareto Front')

# Set labels and title
plt.xlabel('Total Cost')
plt.ylabel('Quality Index')
plt.title('Pareto Front: Total Cost vs Quality Index')

# Format x-axis labels to use scientific notation
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add legend
plt.legend()

# Adjust layout to prevent cutting off labels
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('pareto_front.png', dpi=300)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('f_output.csv')

# Rename columns for clarity
df.columns = ['Total Cost', 'Quality Index']

# Identify the three solutions
economist_choice = df.loc[df['Total Cost'].idxmin()]
environmentalist_choice = df.loc[df['Quality Index'].idxmin()]

# Normalize the data for balanced solution calculation
df_norm = (df - df.min()) / (df.max() - df.min())
df_norm['distance'] = np.sqrt(df_norm['Total Cost']**2 + df_norm['Quality Index']**2)
balanced_choice = df.loc[df_norm['distance'].idxmin()]

# Create the plot
plt.figure(figsize=(14, 10))
plt.scatter(df['Total Cost'], df['Quality Index'], color='blue', alpha=0.6, label='All Solutions')

# Highlight the three chosen solutions and add annotations
solutions = [
    (economist_choice, 'red', 'Economist Choice', (10, 10)),
    (environmentalist_choice, 'green', 'Environmentalist Choice', (10, -10)),
    (balanced_choice, 'purple', 'Balanced Choice', (-10, 10))
]

for solution, color, label, offset in solutions:
    plt.scatter(solution['Total Cost'], solution['Quality Index'], color=color, s=100, label=label)
    plt.annotate(f"Cost: {solution['Total Cost']:.2f}\nQI: {solution['Quality Index']:.6f}",
                 (solution['Total Cost'], solution['Quality Index']),
                 textcoords="offset points",
                 xytext=offset,
                 ha='left',
                 va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                 arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.xlabel('Total Cost')
plt.ylabel('Quality Index')
plt.title('Pareto Front with Highlighted Solutions')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# Adjust layout and display
plt.tight_layout()
plt.savefig('pareto_front_with_choices_and_values.png', dpi=300)
plt.show()

# Print the values of the chosen solutions
for solution, _, label, _ in solutions:
    print(f"\n{label}:")
    print(f"Total Cost: {solution['Total Cost']:.2f}")
    print(f"Quality Index: {solution['Quality Index']:.6f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('f_output.csv')

# Rename columns for clarity
df.columns = ['Total Cost', 'Quality Index']

# Identify the three solutions
economist_choice = df.loc[df['Total Cost'].idxmin()]
environmentalist_choice = df.loc[df['Quality Index'].idxmin()]

# Normalize the data for balanced solution calculation
df_norm = (df - df.min()) / (df.max() - df.min())
df_norm['distance'] = np.sqrt(df_norm['Total Cost']**2 + df_norm['Quality Index']**2)
balanced_choice = df.loc[df_norm['distance'].idxmin()]

# Create the plot
plt.figure(figsize=(14, 10))
plt.scatter(df['Total Cost'], df['Quality Index'], color='blue', alpha=0.6, label='All Solutions')

# Highlight the three chosen solutions and add annotations
solutions = [
    (economist_choice, 'red', 'Economist Choice', (20, 20)),
    (environmentalist_choice, 'green', 'Environmentalist Choice', (20, -20)),
    (balanced_choice, 'purple', 'Balanced Choice', (-20, 20))
]

# Get the axis limits to ensure annotations stay within bounds
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()

for solution, color, label, offset in solutions:
    x, y = solution['Total Cost'], solution['Quality Index']
    # Adjust offsets if needed to stay within axis limits
    offset_x, offset_y = offset
    if x + offset_x > x_max:
        offset_x = x_max - x - 10
    if x + offset_x < x_min:
        offset_x = x_min - x + 10
    if y + offset_y > y_max:
        offset_y = y_max - y - 10
    if y + offset_y < y_min:
        offset_y = y_min - y + 10
    # Create annotation with adjusted offsets
    plt.scatter(x, y, color=color, s=100, label=label)
    plt.annotate(f"Cost: {x:.2f}\nQI: {y:.6f}",
                 (x, y),
                 textcoords="offset points",
                 xytext=(offset_x, offset_y),
                 ha='left',
                 va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                 arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

plt.xlabel('Total Cost')
plt.ylabel('Quality Index')
plt.title('Pareto Front with Highlighted Solutions')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# Adjust layout and display
plt.tight_layout()
plt.savefig('pareto_front_with_choices_and_values.png', dpi=300)
plt.show()

# Print the values of the chosen solutions
for solution, _, label, _ in solutions:
    print(f"\n{label}:")
    print(f"Total Cost: {solution['Total Cost']:.2f}")
    print(f"Quality Index: {solution['Quality Index']:.6f}")
```
